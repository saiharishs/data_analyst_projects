```python
# YouTube API key
API_KEY = 'AIzaSyC6-Sn_t8UcYU1A3vFvSSjYUfr_nUyOdco'
#'AIzaSyBq-0ODqBhd8xy0h6_0uTEkcJRZpSPzlTw'

# YouTube API client
youtube = build('youtube', 'v3', developerKey=API_KEY)

# Mapping dictionary from categoryId to Category Name
category_mapping = {
    '1': 'Film & Animation',
    '2': 'Autos & Vehicles',
    '10': 'Music',
    '15': 'Pets & Animals',
    '17': 'Sports',
    '19': 'Travel & Events',
    '20': 'Gaming',
    '22': 'People & Blogs',
    '23': 'Comedy',
    '24': 'Entertainment',
    '25': 'News & Politics',
    '26': 'Howto & Style',
    '27': 'Education',
    '28': 'Science & Technology',
    '29': 'Nonprofits & Activism',
    '30': 'Movies',
    '31': 'Anime/Animation',
    '32': 'Action/Adventure',
    '33': 'Classics',
    '34': 'Comedy',
    '35': 'Documentary',
    '36': 'Drama',
    '37': 'Family',
    '38': 'Foreign',
    '39': 'Horror',
    '40': 'Sci-Fi/Fantasy',
    '41': 'Thriller',
    '42': 'Shorts',
    '43': 'Shows',
    '44': 'Trailers'
}

def get_channel_data(channel_id):
    response = youtube.channels().list(
        part='snippet,statistics',
        id=channel_id
    ).execute()

    # Check if the response contains 'items' key
    if 'items' in response:
        data = response['items'][0]

        channel_name = data['snippet']['title']
        country = data['snippet'].get('country', 'Unknown')
        description = data['snippet']['description']
        total_views = data['statistics']['viewCount']
        total_subscribers = data['statistics']['subscriberCount']
        video_count = data['statistics']['videoCount']
        category_id = data['snippet'].get('categoryId', 'Unknown'),
        # Map categoryId to category name using the mapping dictionary
        category_name = category_mapping.get(category_id, 'Unknown Category')

        return {
            'Channel Name': channel_name,
            'Country': country,
            'Description': description,
            'Total Views': total_views,
            'Total Subscribers': total_subscribers,
            'Video Count': video_count,
            'Category Name': category_name,
            'Channel ID': channel_id,
            'Channel URL': f'https://www.youtube.com/channel/{channel_id}'

        }
    else:
        print(f"No data found for channel ID: {channel_id}")  # Print a message if 'items' key is missing
        return None  # Return None to indicate no data


def get_videos(channel_id, max_results=50):
  # Try to fetch videos using the provided channel_id
    try:
        response = youtube.search().list(
            channelId=channel_id,
            part='snippet',
            order='date',
            maxResults=max_results
        ).execute()
    # If the request fails due to an invalid channel ID, print an error message
    except HttpError as e:
        print(f"Error fetching videos for channel ID: {channel_id}. Error: {e}")
        return [] # Return an empty list to avoid further errors

    videos = []

    for item in response.get('items', []): # Handle cases where 'items' might not be present
        video_id = item['id'].get('videoId')
        if video_id:
            video_type = 'Video'
        else:
            continue

        video_url = f'https://www.youtube.com/watch?v={video_id}'

        videos.append({
            'Video URL': video_url,
            'Video Type': video_type
        })

    return videos


def get_views(video_id, days):
    now = datetime.now()
    start_time = (now - timedelta(days=days)).isoformat("T") + "Z"

    response = youtube.videos().list(
        part='statistics',
        id=video_id
    ).execute()

    views = response['items'][0]['statistics'].get('viewCount', 0)

    return views

channel_ids = [
    "UC_aEa8K-EOJ3D6gOs7HcyNg",
    "UCS5Oz6CHmeoF7vSad0qqXfw",
    "UCVtFOytbRpEvzLjvqGG5gxQ",
    "UC_8PAD0Qmi6_gpe77S1Atgg",
    "UC0DZmkupLYwc0yDsfocLh0A",
    "UCcartHVtvAUzfajflyeT_Gg",
    "UC-J-KZfRV8c13fOCkhXdLiQ",
    "UC6D1L2vxEAg_Vi0JSxMBDgA",
    "UCYVinkwSX7szARULgYpvhLw",
    "UCHwBjRoGO1V-H8QYXjSVHkQ",
    "UCiMhD4jzUqG-IgPzUmmytRQ",
    "UCIvqWqjC4aLvp7zI8hzARrA",
    "UCGmnsW623G1r-Chmo5RB4Yw",
    "UCKvn9VBLAiLiYL4FFJHri6g",
    "UC16niRr50-MSBwiO3YDb3RA",
    "UCCj956IF62FbT7Gouszaj9w",
    "UCBq5hp6DvwWam2XOXw6lYJA",
    "UCLeuoGy_hUDTBf5Hk0ynrpQ",
    "UCp68_FLety0O-n9QU6phsgw",
    "UCMiJRAwDNSNzuYeN2uWa0pA",
    "UC5p_l5ZeB_wGjO_yDXwiqvw",
    "UCBwSufNse8VMBvQM_rCSvgQ",
    "UC6sSkkemzPjmrzS0Y0V_2zw",
    "UC7Gf2tZ8coTX2ckTPgn62iQ",
    "UC-0D1RlSzcvBoODCKSoLl2g",
    "UCHdMK5Ef2El8KbD3L_WgANg",
    "UC4pEp0mo54WhEBCTYvjVDJA",
    "UCwTRjvjVge51X-ILJ4i22ew",
    "UCelk6aHijZq-GJBBB9YpReA",
    "UCEwIUtFBhaI2L2PuKv0KL2g",
    "UC1ieoHqKW-yYgDhLHIcx28w",
    "UCfIXdjDQH9Fau7y99_Orpjw",
    "UCoGForHQAmk0_9rI9CABeYw",
    "UChUHOEe7zgYmrv7a_LlpNng",
    "UCEY1ejsweY4DgMwOVJeEaBA",
    "UCz4tgANd4yy8Oe0iXCdSWfA",
    "UCS7H8U-n5mINVJjJsaRtGHg",
    "UC977n5nXRn8MDuGQjwxkOLw",
    "UChFur_NwVSbUozOcF_F2kMg",
    "UCjOl2AUblVmg2rA_cRgZkFg",
    "UC1QqdFv5e9QSEi0eYdfwuqw",
    "UC-FQUIVQ-bZiefzBiQAa8Fw",
    "UCR9Gcq0CMm6YgTzsDxAxjOQ",
    "UCDBCCMlFD-hArPbWv38FINA",
    "UCvwgF_0NOZe2vN4Q3g1bY-A",
    "UCHFA6Nt11kE_8OWIya0SQDQ",
    "UCUhFaUpnq31m6TNX2VKVSVA",
    "UC5Y1pPVNm-odhiubuRO1D6A",
    "UCOYWgypDktXdb-HfZnSMK6A",
    "UCH-_hzb2ILSCo9ftVSnrCIQ",
    "UCzTlXb7ivVzuFlugVCv3Kvg",
    "UCc4K7bAqpdBP8jh1j9XZAww",
    "UC_wITz-Ze4JFc3WLQA1n3Hw",
    "UCRlICXvO4XR4HMeEB9JjDlA",
    "UCcmpeVbSSQlZRvHfdC-CRwg",
    "UC_vXQAIUPGZdFy5I9Y9oKgA",
    "UC_SY5JxKNyNlyEXvI2v9B8g",
    "UCa_MZBj2EeP6d6u7YZAQddA",
    "UC5GoG9W_9cZywyoq4zZx1FQ",
    "UCtH6N6v8eZFBzFmAn_gj4dQ",
    "UCy-tnaK0F-RttJ3J_2xQU2Q",
    "UC2Q3qzIu5IGXt0I3L6BtPAQ",
    "UCO8DQrSp5yEP937qNqTooOw",
    "UC_LI8JbzXyRoIb_ei2FQqBQ",
    "UC4ELwkuEg4AcKp8eSE-9m-g",
    "UC2C_jShtL725hvbm1arSV9w",
    "UCgqfY0vqyKpWg4H2GK-9uFA",
    "UCP-_q0LkHBy-5ZpzWvA6TIQ",
    "UC-VDfQNL-S44eZscGlXI14g",
    "UCakgsb0w7QB0VHdnCc-OVEA",
    "UCu4XcDBdnZkV6-5z2f16M0g",
    "UCO66lK14MYiXlu0A6pQ4ouA",
    "UCqTwKvjbTENZDGbz2si47ag",
    "UCLk9wddjT0F-HILLRFXodtg",
    "UCbmWOFTbZoDLWqG1rvFnJ0g",
    "UCoMdktPbSTixAyNGwb-UYkQ",
    "UCvxfEIG3PHpgM0TMJJ_SH-w",
    "UCRlICXvO4XR4HMeEB9JjDlA",
    "UCHhfSXoDG6gSgpOvLH4wrRw",
    "UCHaHD477h-FeBbVh9Sh7syA",
    "UCkzCjdRMrW2vXLx8mvPVLdQ",
    "UC0h07r_UgTD0Tc-Dn5XLX3g",
    "UCaFUrR3oSxOl5Y9y6tvLTEg",
    "UChntGq8THlUokhc1tT-M2wA",
    "UChoQQRDf-zU1h6_YjKkWvdw",
    "UCbmWOFTbZoDLWqG1rvFnJ0g",
    "UCoMdktPbSTixAyNGwb-UYkQ",
    "UCvxfEIG3PHpgM0TMJJ_SH-w"
]

data = []

for channel_id in channel_ids:
    channel_data = get_channel_data(channel_id)
    videos = get_videos(channel_id, max_results=5)

    for video in videos:
        video_id = video['Video URL'].split('v=')[1]
        last_month_views = get_views(video_id, 30)
        last_week_views = get_views(video_id, 7)

        data.append({
            'Unique ID': channel_id,
            'Channel Name': channel_data['Channel Name'],
            'Country': channel_data['Country'],
            'Last Video URL': video['Video URL'],
            'Total Views': channel_data['Total Views'],
            'Video Type': video['Video Type'],
            'Last 1 Month Views': last_month_views,
            'Last 1 Week Views': last_week_views,
            'Total Subscribers': channel_data['Total Subscribers'],
            'Channel Description': channel_data['Description'],
            'Total View Count': channel_data['Total Views'],
            'Video Count': channel_data['Video Count'],
            'Category Name': channel_data['Category Name'],
            'Channel URL': channel_data['Channel URL']
        })

df = pd.DataFrame(data)
df.to_csv('top_100_uk_youtubers.csv', index=False)

print("Data collection complete. File saved as top_100_uk_youtubers.csv")
```
