1. Find all the unique 'Wind Speed' values in the data.

2. Find the number of times when the 'Weather is exactly Clear'.

```python
# Find the number of times 'Weather is exactly Clear'
clear_weather_count = weather_df[weather_df['Weather'] == 'Clear'].shape[0]

print(f"The number of times when the weather is exactly 'Clear': {clear_weather_count}")

```

```python
# displaying the some values where weather is clear
weather_df.groupby('Weather').get_group('Clear').head(1)
```

5) Rename the column name 'Weather' of the dataframe to 'Weather Condition'.


```python
# renaming the column
weather_df.rename(columns={'Weather': 'Weather_Condition'}, inplace=True)
list(weather_df.columns)
```

6) What is the mean 'Visibility' ?


```python
# mean,std,variance for Visiblity column
print('MEAN - VISIBILITY -> ',weather_df.Visibility_km.mean())
print('STD - VISIBILITY -> ',weather_df.Visibility_km.std())
print('VAR - VISIBILITY -> ',weather_df.Visibility_km.var())
```

7) Find all instances when 'Snow' was recorded.

```python
# Filter the DataFrame to find instances of 'Snow'
snow_instances = weather_df[weather_df['Weather_Condition'] == 'Snow']

# Count the instances of 'Snow'
snow_count = snow_instances.shape[0]

print(f"Instances of 'Snow':\n{snow_instances}")
print(f"Number of times 'Snow' was recorded: {snow_count}")
```

```python
# using string function to fetch the instances - columns contains word -> Snow 
# str.contains
snow_df = weather_df[weather_df['Weather_Condition'].str.contains(r'\bSnow\b', regex=True)]
print('Number of SNOW/SNOWY/SNOWY SHOWERS condition weather reported - ',snow_df.shape[0])

# using string function to fetch the instances - columns contains word exactly -> Snow 
# str.exact - match(r'^Snow$')]
snow_df = weather_df[weather_df['Weather_Condition'].str.match(r'^Snow$')]
print('Number of exactly SNOW condition weather reported - ',snow_df.shape[0])
```

7) Find all instances when 'Wind Speed is above 24' and 'Visibility is 25'.

```python
# displaying the Weather report where visiblity = 25 and wind speed > 24
weather_df[(weather_df['Visibility_km'] == 25) & (weather_df['Wind Speed_km/h'] > 24)]
```

8) What is the Mean value of each column against each 'Weather Conditon' ?

```python
# Select numeric columns only
numeric_cols = weather_df.select_dtypes(include='number')

# Group by 'Weather' and calculate the mean of each numeric column
mean_values = numeric_cols.groupby(weather_df['Weather_Condition']).mean()

# Display the mean values
print("\nMean Values of Numeric Columns by Weather Condition:")
mean_values
```

9) What is the Minimum & Maximum & sum value of each column against each 'Weather Conditon' ?

```python
# Group by 'Weather' and calculate min and max of each numeric column
min_max_values = numeric_cols.groupby(weather_df['Weather_Condition']).agg(['min', 'max', 'sum'])

print("\nMinimum and Maximum Values of Numeric Columns by Weather Condition:")
min_max_values
```
