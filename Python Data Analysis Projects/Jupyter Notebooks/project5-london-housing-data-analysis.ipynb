```python
# convert date time column into pd date time type
house_df['date'] = pd.to_datetime(house_df['date'])
```

```python
# check data types
house_df.dtypes
```

**(B.1) Add a new column ''year'' in the dataframe, which contains years only.**

```python
house_df['Year'] = house_df.date.dt.year
```

```python
house_df.groupby('Year').nunique()
```

**(B.2) Add a new column ''month'' as 2nd column in the dataframe, which contains month only.**

```python
house_df['Month'] = house_df.date.dt.month
house_df.head()
```

```python
house_df.groupby(['Month']).first()
```

```python
df1 = house_df[house_df.area =='england']
df1.groupby('Year').average_price.max()
```

```python
df1 = house_df[house_df.area =='england']
df1.groupby('Year').average_price.min()
```

```python
df1 = house_df[house_df.area =='england']
df1.groupby('Year').average_price.mean()
```

```python
house_df.groupby('area').no_of_crimes.min().sort_values(ascending = False)
```

```python
# (G) Show the total count of records of each area, where average price is less than 100000.
house_df[house_df.average_price < 100000].area.value_counts()
```
