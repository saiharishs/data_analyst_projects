**DATA LOADING**

```python
covid_df = pd.read_csv('/kaggle/input/datasets-py/Covid_19_data.csv')
covid_df.head()
```

**DATA PREVIEWING**

```python
# check null values present in the column
covid_df.isnull().sum()
```

```python
# Remove all the records where Confirmed Cases is Less Than 10.
covid_df = covid_df[~(covid_df.Confirmed < 10)]
covid_df.Confirmed.value_counts()
```

```python
#In which Region, maximum number of Confirmed cases were recorded ?
covid_df.groupby('Region').Confirmed.sum().sort_values(ascending = False).head(15)
```

```python
# In which Region, minimum number of Deaths cases were recorded ?
covid_df.groupby('Region').Deaths.sum().sort_values().head(30)
```

```python
# How many Confirmed , Deaths & Recovered cases were reported from India till 29 April 2020 ?
covid_df[covid_df['Region'] == 'India'].loc[:,['Region','Confirmed','Deaths']]
```

```python
# sort data by column wise 
# sort the Recovered case by desending order

covid_df.sort_values(by = ['Recovered'], ascending = False).loc[:,['Region','Confirmed','Deaths', 'Recovered']].head(10)
```
