```python
df[df['price'] == 'Free']
```

```python
#  Show all the courses which are Paid.
df[df.is_paid == True]    
```

```python
# Top selling courses
df.sort_values(by = 'num_subscribers', ascending = False)[:10]
```

```python
# Top Least courses
df.sort_values(by = 'num_subscribers', ascending = True)[:10]
```

```python
# Show all courses of Graphic Design where the price is below 100 ?
df[(df.subject == 'Graphic Design') & (df.price > '100')]
```

```python
# List out all the courses that are related with 'Python'.
len(df[df.course_title.str.contains('Python')])
```

```python
#  What are courses that published in year 2015 ?
df['published_timestamp'] = pd.to_datetime(df.published_timestamp)
```

```python
df['Year'] = df['published_timestamp'].dt.year
```

```python
df[df.Year == 2015]
```

```python
# What are the Max. Number of Subscribers for Each Level of courses ?
df.groupby('level')['num_subscribers'].max()
```
