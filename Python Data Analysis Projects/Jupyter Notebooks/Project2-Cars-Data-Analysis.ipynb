***DATA ANALYSIS CARS DATASET***

**Car Data Analysis**

Car Data Analysis involves examining various attributes of vehicles to uncover trends and insights about their performance, market value, and design specifications. This analysis includes columns such as:

- **Make**: The manufacturer of the car.
- **Model**: The specific name of the car model.
- **Type**: The category or body style of the car (e.g., sedan, SUV).
- **Origin**: The country where the car was manufactured.
- **DriveTrain**: The type of drivetrain (e.g., FWD, RWD, AWD).
- **MSRP**: Manufacturer's Suggested Retail Price.
- **Invoice**: The dealer's cost for the car.
- **EngineSize**: The size of the car's engine in liters.
- **Cylinders**: The number of cylinders in the car's engine.
- **Horsepower**: The power output of the car's engine.
- **MPG_City**: The car's fuel efficiency in the city (miles per gallon).
- **MPG_Highway**: The car's fuel efficiency on the highway.
- **Weight**: The weight of the car.
- **Wheelbase**: The distance between the front and rear axles.
- **Length**: The overall length of the car.

Analyzing this data can help in understanding vehicle performance, identifying market trends, evaluating pricing strategies, and comparing specifications across different car models and makes.


**DATA CLEANING**

1)Find all Null Values in the dataset. If there is any null value in any column, then fill it with the mean of that column.

```python
# null value fill na with mean 
cars_df.isnull().sum()
```

```python
# Fill numeric columns with the mean of their respective columns
for column in cars_df.select_dtypes(include=['float64', 'int64']):
    cars_df[column].fillna(cars_df[column].mean(), inplace=True)

# Fill non-numeric columns with 'Unknown' or a similar placeholder
for column in cars_df.select_dtypes(include=['object']):
    cars_df[column].fillna('Unknown', inplace=True)
```

```python
# checking df after filling mean values
cars_df.isna().sum()
```

2) Check what are the different types of Make are there in our dataset. And, what is the count (occurrence) of each Make in the data ?

3) Show all the records where Origin is Asia or Europe.


```python
# using or operator
cars_df[(cars_df['Origin'] == 'Asia') | (cars_df['Origin'] == 'Europe')]
```

```python
# using is in function 
cars_df[cars_df['Origin'].isin(['Asia', 'Europe'])]
```

```python
cars_df['Origin'].value_counts()
```

4) Remove all the records (rows) where Weight is above 4000.

```python
# using ~ symbol - removing data and keeping only filtered data in the data frame.
cars_df[~(cars_df['Weight'] >= 4000)]
```

```python
# Remove rows where Weight is above 4000
cars_df = cars_df[cars_df['Weight'] <= 4000]
```
